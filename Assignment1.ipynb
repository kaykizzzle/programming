```python
import numpy as np
import matplotlib.pyplot as plt
grid = np.linspace(0,1, num = 50)
log = np.log(grid)
exp = np.exp(grid)
plt.scatter(grid,log)
plt.scatter(grid,exp)
plt.show()

grid = np.linspace(0,1, num = 50)
log = np.log(grid)
exp = np.exp(grid)
plt.scatter(grid,log, label='Natural Log')
plt.scatter(grid,exp, label='Exponential')
plt.xlabel("50 points ranging from 0 to 1")
plt.ylabel("natural log and exponential value of point")
plt.title("Natural Log and Exponential Functions")
plt.legend(loc = 'lower right')
plt.show()

grid2 = np.linspace(-6.5,6.5, num = 100)
sin = np.sin(grid2)
cos = np.cos(grid2)
plt.scatter(grid2,sin)
plt.scatter(grid2,cos)
plt.show()

plt.plot(grid2,sin)
plt.plot(grid2,cos)
plt.show()

plt.plot(grid2,sin, label = "sine")
plt.plot(grid2,cos, label = "cosine")
plt.xlabel("100 points ranging from -6.5 to 6.5")
plt.ylabel("cosine and sine of point")
plt.title("Cosine and Sine Functions")
plt.legend(loc = 'lower left')
plt.show()
```
